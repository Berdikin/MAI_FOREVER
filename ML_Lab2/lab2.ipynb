### Датасет

Для начала необходимо загрузить датасет из первой лабораторной работы:

```python
import pandas as pd
import numpy as np

data = pd.read_csv('mushroom.csv').dropna().drop(columns=['veil-type'])
data
```

```python
cols = data.columns
data[cols] = data[cols].apply(lambda x: pd.Categorical(x).codes)
data
```

Как и в предыдущей лабораторной работе, нам необходимо определить съедобность гриба:

```python
X, Y = data.drop(columns=['class']).to_numpy(), np.array(data['class'])
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.15, random_state=15)
```

Необходимо перенести функцию для вывода метрик Accuracy, Precision и Recall

```python
#positive -- ядовитый, negative -- съедобный, y -- точные значения, y0 -- вычисленные
def print_accuracy(y, y0):
    tp, fp, tn, fn = 0, 0, 0, 0
    
    for i in range(y.shape[0]):
        if y[i] == 0 and y0[i] == 0: tn += 1
        if y[i] == 1 and y0[i] == 0: fn += 1
        if y[i] == 0 and y0[i] == 1: fp += 1
        if y[i] == 1 and y0[i] == 1: tp += 1
        
    print('Accuracy:', (tp + tn) / (tp + tn + fp + fn))
    print('Precision:', tp / (tp + fp))
    print('Recall:', tp / (tp + fn))
```

### Логистическая регрессия

Необходимо написать логистическую функцию:

```python
from math import exp

def logist(x):
    return 1 / (1 + exp(-np.dot(x, lr_weights)))
```

Теперь нам нужно найти для логистической функции веса. Назовём их lr_weights.

```python
lr_weights = np.zeros(Xtrain.shape[1] + 1)
lr_xtrain = np.hstack((Xtrain, np.ones((Xtrain.shape[0], 1)))) #добавляем единицы в конец каждой выборки

#процесс обучения с шагом 0.1
for repeat in range(100):
    for i in range(lr_xtrain.shape[0]):
        x = lr_xtrain[i]
        y = Ytrain[i]
        lr_weights = lr_weights + 0.05 * np.dot(y - logist(x), x)
```

И, наконец, проверяем, насколько точна полученная нами регрессия:

```python
lr_y0 = np.zeros(Ytest.shape[0])

for i in range(Ytest.shape[0]):
    if logist(np.append(Xtest[i], 1)) < 0.5:
        lr_y0[i] = 0
    else:
        lr_y0[i] = 1

print_accuracy(Ytest, lr_y0)
```

А насколько совпадает наша точность с реализацией sklearn?

```python
from sklearn.linear_model import LogisticRegression
sk_lr = LogisticRegression(max_iter=500)
sk_lr.fit(Xtrain, Ytrain)

print_accuracy(Ytest, sk_lr.predict(Xtest))
```

Точности поразительно схожи. И наша точность довольно высока, т.е. можно сделать вывод о том, что съедобность гриба легко определяется простой линейной функцией от его признаков.

### Дерево решений

Для описания вершины дерева лучше всего создать класс:

```python
class Node:
    def __init__(self, index, value):
        self.index = index
        self.value = value
        self.left = None
        self.right = None
```

По алгоритму ID3 построим дерево решений: 

```python
def Gini(X, Y, i, v):
    #найдём число пар, принадлежащих одинаковым классам и одинаковым разбиениям
    YL = Y[X[::, i] <= v]
    YR = Y[X[::, i] > v]
    
    yl1 = YL.sum()
    yl0 = YL.shape[0] - yl1
    
    yr1 = YR.sum()
    yr0 = YR.shape[0] - yr1
    
    return (yl0 * (yl0 + 1) + yl1 * (yl1 + 1) + yr0 * (yr0 + 1) + yr1 * (yr1 + 1))

def BestGini(X, Y):
    index = 0
    value = 0
    gini = 0
    
    for i in range(X.shape[1]):
        for v in set(X[::, i]):
            g = Gini(X, Y, i, v)
            
            if g < gini or gini == 0:
                gini = g
                index = i
                value = v
    
    return index, value

def ID3(X, Y, depth):
    #все вершины одного класса
    if Y.min() == Y.max():
        return Node(-1, Y[0])
    #максимальная глубина достигнута -- самый распространённый класс
    if depth == 0:
        return Node(-1, round(Y.mean()))
    
    Node(-1, round(Y.mean()))
    #находим разбиение по критерию Джини
    index, value = BestGini(X, Y)
    
    #разбиваем
    XL = X[X[::, index] <= value]
    XR = X[X[::, index] > value]
    YL = Y[X[::, index] <= value]
    YR = Y[X[::, index] > value]
    
    #одно из разбиений пустое -- самый распространённый класс
    if XL.shape[0] == 0 or XR.shape[0] == 0:
        return Node(-1, round(Y.mean()))
    
    #создаём вершину и её поддеревья
    n = Node(index, value)
    n.left = ID3(XL, YL, depth - 1)
    n.right = ID3(XR, YR, depth - 1)
    return n

#вычислить значение по дереву
def getClass(tree, x):
    if tree.index == -1:
        return tree.value
    
    if x[tree.index] <= tree.value:
        return getClass(tree.left, x)
    else:
        return getClass(tree.right, x)
```

```python
tree = ID3(Xtrain, Ytrain, 15)

tree_y0 = np.zeros(Ytest.shape[0])

for i in range(Ytest.shape[0]):
    tree_y0[i] = getClass(tree, Xtest[i])

print_accuracy(Ytest, tree_y0)
```

### SVM

Формула для SVM похожа на формулу для логистической регрессии и выглядит следующим образом: $$f(x)=sign(w\cdot x-b)$$

Нам необходимо найти вектор w и число b:

```python
svm_w = np.zeros(Xtrain.shape[1])
svm_b = 0

def f(x):
    return 1 if np.dot(svm_w, x) - svm_b > 0 else -1

#преобразуем Ytrain из [0, 1] в [-1, 1]
svm_y = 2 * Ytrain - 1

#обучение
rate = 0.01
lambd = 0.02

for repeat in range(250):
    for i in range(Xtrain.shape[0]):
        x = Xtrain[i]
        if 1 - svm_y[i] * (np.dot(svm_w, x) - svm_b) >= 0:
            svm_w = svm_w * (1 - rate * lambd) + rate * svm_y[i] * x
            svm_b = svm_b - rate * svm_y[i]
        else:
            svm_w = svm_w * (1 - rate * lambd)
```

```python
svm_y0 = np.zeros(Ytest.shape[0])

for i in range(Ytest.shape[0]):
    if f(Xtest[i]) < 0:
        lr_y0[i] = 0
    else:
        lr_y0[i] = 1

print_accuracy(Ytest, lr_y0)
```
